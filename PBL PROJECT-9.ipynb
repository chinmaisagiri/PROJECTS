```python
labelencoder=preprocessing.LabelEncoder()
df_encoded=df.apply(labelencoder.fit_transform)
df_encoded
```

```python
fea=list(zip(df_encoded['Weather'],df_encoded['Temp']))
fea
```

```python
model = GaussianNB()
model.fit(fea,df_encoded['Play'])
model.predict(fea)
```

```python
model.score(fea,df_encoded['Play'])
```
